# Linear Regression Analysis - Profit Drivers
Detailed analysis of profit drivers using linear regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

from src.data_prep import prepare_data
from src.features import run_feature_engineering
from src.models.linear_model import run_linear_model, get_coefficients
```

## Load and Prepare Data

```python
df = prepare_data()
df = run_feature_engineering(df)
df.head()
```

## Train Linear Model

```python
model, X, y = run_linear_model(df)
print(model.summary())
```

## Coefficients Analysis

```python
coef_df = get_coefficients(model)
coef_df
```

### Interpretation Guide
- Coefficient > 0: increases profit
- Coefficient < 0: decreases profit
- p-value < 0.05: statistically significant

## Business Questions

```python
price_coef = coef_df[coef_df['feature']=='price']
freight_coef = coef_df[coef_df['feature']=='freight_value']
discount_coef = coef_df[coef_df['feature']=='discount_ratio']

price_coef, freight_coef, discount_coef
```

### Interpretation
- Price: positive coefficient suggests higher prices increase profit
- Freight: negative coefficient suggests logistics reduces margin
- Discount: negative suggests promotions may harm profitability

## Model Performance

```python
print('R-squared:', model.rsquared)
```

## Impact Visualization

```python
coef_df.set_index('feature')['coefficient'].plot(kind='bar')
plt.title('Feature Impact on Profit')
plt.ylabel('Coefficient')
plt.show()
```

## Residual Analysis

```python
residuals = model.resid
plt.hist(residuals, bins=50)
plt.title('Residual Distribution')
plt.show()
```

## Final Business Insights
- Pricing is a primary lever for profit optimization
- Logistics costs significantly reduce margins
- Promotions should be applied selectively
- Model provides a strong baseline for decision-making
